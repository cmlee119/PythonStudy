```python
import collections as std

T = int(input())
for _ in range(T):
    N = int(input())
    
    queue = std.deque([])
    
    for _ in range(N):
        V = int(input())
        if V == -1:
            print(queue.popleft())
        else:
            queue.append(V)
```

```python
import heapq

T = int(input())
for _ in range(T):
    N = int(input())
    
    hq = []
    
    for _ in range(N):
        V = int(input())
        if V == -1:
            print(heapq.heappop(hq))
        else:
            heapq.heappush(hq, V)
```

```python
def hanoi(n, x, y, z):
    if n == 1:
        print('%c -> %c' % (x, y))
    else:
        hanoi(n-1, x, z, y)
        hanoi(1, x, y, z)
        hanoi(n-1, z, y, x)


T = int(input())
for _ in range(T):
    N = int(input())
    
    hanoi(N, 'A', 'C', 'B')
```

```python
import collections as std

T = int(input())
for _ in range(T):
    N = int(input())
    listNum = list(map(int, input().split()))

    listQueue = []
    for _ in range(N):
        listQueue.append(std.deque([]))
    
    for num in listNum:  
        bFind = False
        for i in range(N):
            queue = listQueue[i]
            if len(queue) != 0 and num == queue[0]:
                #print('pop', i, num)
                bFind = True
                V = queue.popleft()
                if i + 1 < N - 1:
                    #print('reappend', i + 1, num)
                    listQueue[i + 1].append(V)
                break
        
        if bFind == False:
            #print('firstappend', i, num)
            listQueue[0].append(num)

    total = 0     
    for queue in listQueue:
        total += sum(queue)
    
    if total == 0:
        print('NO')
    else:
        print('YES')
        
```
